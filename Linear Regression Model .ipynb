# Importing Libraries and Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("dark_background")
```

```python
data = pd.read_csv('Transformed_Housing_Data2.csv')
data.head()
```

# Scaling The Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Y = data['Sale_Price']
X = scaler.fit_transform(data.drop(columns = ['Sale_Price']))
x = pd.DataFrame(data=X, columns = data.drop(columns = ['Sale_Price']).columns)
x.head()
```

# Checking and removing multicollinearity

```python
x.corr()
```

```python
# pair of independent variables with correlation greater than 0.5
k = x.corr()
# list comprehension
z = [(str(i),str(j)) for i in k.columns for j in k.columns if(k.loc[i,j]> abs(0.5))& (i!=j)]
z, len(z)
```

```python
#removing them using vif
from statsmodels.stats.outliers_influence import variance_inflation_factor as Vif
vif_data = x
# calculating vif for  every column
VIF = pd.Series([Vif(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
# there are 7 columns whose VIF > 5
VIF[VIF == VIF.max()].index[0]
```

```python
def mc_remover(data):
    vif = pd.Series([Vif(data.values,i) for i in range(data.shape[1])],index = data.columns)
    if vif.max() > 5:
        print(vif[vif == vif.max()].index[0], 'has been removed')
        data = data.drop(columns = [vif[vif == vif.max()].index[0]])
        return data
    else:
        print('No multicollinearity present anymore')
        return data
```

```python
for i in range(7):
    vif_data = mc_remover(vif_data)
vif_data.head()    
```

# Remaining Columns

```python
# calcuting VIF for remaining columns
VIF = pd.Series([Vif(vif_data.values,i) for i in range(vif_data.shape[1])],index = vif_data.columns)
VIF, len(VIF)
```

# Training the model

```python
x = vif_data
y = data['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 101)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Implementing Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(normalize = True)
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
# Generating predictions over test set
predictions = lr. predict(x_test)
lr.score(x_test, y_test)
```

# Making Residuals Plots

```python
residuals = predictions - y_test
residual_table = pd.DataFrame({'residuals': residuals, 'predictions': predictions})
residual_table = residual_table.sort_values(by = 'predictions')
```

```python
z = [i for i in range(int(residual_table['predictions'].max()))]
k = [0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi = 130, figsize = (17,7))
plt.scatter(residual_table['predictions'], residual_table['residuals'],color='red',s= 2)
plt.plot(z,k,color = 'green', linewidth = 3, label = 'regression line')
plt.ylim(800000,-800000)
plt.xlabel('fitted points(ordered by predictions)')
plt.ylabel('residuals')
plt.title('residual plot')
plt.legend()
plt.show()
```

# Distribution of Errors

```python
plt.figure(dpi = 100, figsize = (10,7) )
plt.hist(residual_table['residuals'], color = 'red', bins = 200)
plt.xlabel('residuals')
plt.ylabel('frequency')
plt.title('Distribution of residuals')
plt.show()
```

# Visualizing Coefficients to interpret the model result

```python
# Model Coefficients

coefficient_table = pd.DataFrame({'columns':x_train.columns, 'coefficients': lr.coef_})
coefficient_table = coefficient_table.sort_values(by = 'coefficients')
plt.figure(figsize = (8,6), dpi = 120)
x = coefficient_table['columns']
y = coefficient_table['coefficients']
plt.bar(x,y)
plt.xlabel('columns')
plt.ylabel('coefficients')
plt.title('Normalised Coefficient plot')
plt.show()
```
