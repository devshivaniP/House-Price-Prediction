# Low Variance Removal

```python
data1 = pd.read_csv("Raw_Housing_Prices3.csv")
```

```python
data1.head()
```

```python
data1.info()
```

```python
numerical_data= data1.drop(columns =['ID','Date House was Sold','Sale Price',
                                    'Waterfront View','No of Times Visited',
                                    'Condition of the House','Overall Grade','Zipcode'])
numerical_data.info()
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()
norm_data = norm.fit_transform(numerical_data)
norm_data = pd.DataFrame(norm_data, columns = numerical_data.columns)
norm_data.head()
```

```python
tmp = norm_data.var()
tmp
```

```python
T = 1e-04
for i in tmp:
    if i<T:
        norm_data.drop(columns = [tmp[tmp==i].index[0]],inplace = True)
norm_data.var()        
        
```
