```python
import pandas
import numpy as np
```

```python
Raw_Housing_Data2 = pandas.read_csv("Raw_Housing_Prices3.csv")
```

```python
Raw_Housing_Data2.info()
```

```python
# to change data type
Raw_Housing_Data2['Zipcode'] = Raw_Housing_Data2['Zipcode'].astype(object)
Raw_Housing_Data2.dtypes
```

```python
# to chnage data type with values
Raw_Housing_Data2['No of Times Visited'].unique()
```

```python
mapping = { 'None' : '0',
          'Once': '1',
          'Twice' : '2',
          'Thrice' : '3',
          'Four' : '4',
          }
Raw_Housing_Data2['No of Times Visited'] = Raw_Housing_Data2['No of Times Visited'].map(mapping)
```

```python
Raw_Housing_Data2.head(10)
```

```python
# to create variables from an existing variable
Raw_Housing_Data2['Ever Renovated'] = np.where(Raw_Housing_Data2['Renovated Year'] == 0, 'No','Yes')
Raw_Housing_Data2
```

```python
Raw_Housing_Data2['Purchase Year'] = pandas.DatetimeIndex(Raw_Housing_Data2['Date House was Sold']).year
Raw_Housing_Data2['Year Since Renovation'] = np.where(Raw_Housing_Data2['Ever Renovated'] == 'Yes', abs(Raw_Housing_Data2['Purchase Year'] - Raw_Housing_Data2['Renovated Year']),0)
```

```python
Raw_Housing_Data2.head(10)
```

```python
Raw_Housing_Data2.drop(columns = ['Date House was Sold','Purchase Year','Renovated Year'], inplace = True)
```

```python
Raw_Housing_Data2.head(10)
```

```python
Raw_Housing_Data2.to_csv('Transformed Data', index = False)
```
